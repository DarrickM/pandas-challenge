```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
schoolData = pd.read_csv(school_data_to_load)
studentData = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
schoolDF = pd.merge(studentData, schoolData, how="left", on=["school_name", "school_name"])
schoolDF.head()
```

## District Summary

```python
# Calculate the total number of unique schools
schoolCount = schoolDF['school_name'].unique()
schoolCount
```

```python
# Calculate the total number of students
studentCount = len(schoolDF['student_name'])
studentCount
```

```python
# Calculate the total budget
totalBudget = sum(schoolDF['budget'].unique())
totalBudget
```

```python
# Calculate the average (mean) math score
averageMathScore = round(schoolDF['math_score'].mean(), 2)
averageMathScore
```

```python
# Calculate the average (mean) reading score
averageReadingScore = round(schoolDF['reading_score'].mean(), 2)
averageReadingScore
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passingMathCount = schoolDF[(schoolDF["math_score"] >= 70)].count()["student_name"]
passingMathPercentage = round(passingMathCount / float(studentCount) * 100, 2)
passingMathPercentage
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passingReadingCount = schoolDF[(schoolDF["reading_score"] >= 70)].count()["student_name"]
passingReadingPercentage = round(passingReadingCount / float(studentCount) * 100, 2)
passingReadingPercentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passingMathReadingCount = schoolDF[
    (schoolDF["math_score"] >= 70) & (schoolDF["reading_score"] >= 70)].count()["student_name"]
overallPassingRate = round(passingMathReadingCount /  float(studentCount) * 100, 2)
overallPassingRate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
districtSummary = pd.DataFrame()
districtSummary[""] = [""]
districtSummary["Total Schools"] = len(schoolCount)
districtSummary["Total Students"] = studentCount
districtSummary["Total Budget"] = totalBudget
districtSummary["Average Math Score"] = averageMathScore
districtSummary["Average Reading Score"] = averageReadingScore
districtSummary["% Passing Math"] = passingMathPercentage
districtSummary["% Passing Reading"] = passingReadingPercentage
districtSummary["% Overall Passing"] = overallPassingRate


# Formatting

districtSummary["Total Students"] = districtSummary["Total Students"].map("{:,}".format)
districtSummary["Total Budget"] = districtSummary["Total Budget"].map("${:,.2f}".format)


# Display the DataFrame
districtSummary
```

```python
# Calculate the number of students per school with math scores of 70 or higher
# Calculate the number of students per school with reading scores of 70 or higher
students_passing_math = len(schoolDF.loc[schoolDF['math_score'] >= 70])
school_students_passing_math = []
for x in range(15):
    school_students_passing_math.append(len(schoolDF.loc[(schoolDF['math_score'] >= 70) & 
                     (schoolDF["school_name"] == schoolData2["school_name"][x])]))

students_passing_reading = len(schoolDF.loc[schoolDF['reading_score'] >= 70])
school_students_passing_reading = []
for x in range(15):
    school_students_passing_reading.append(len(schoolDF.loc[(schoolDF['reading_score'] >= 70) & 
                     (schoolDF["school_name"] == schoolData2["school_name"][x])]))
school_students_passing_reading
schoolData2["passed reading"] = school_students_passing_reading
schoolData2["passed math"] = school_students_passing_math
schoolData2
```

```python
# Use the provided code to calculate the number of students per school that passed both
# math and reading with scores of 70 or higher
students_passing_math_and_reading = []
for x in range(15):
    students_passing_math_and_reading.append(len(schoolDF.loc[(schoolDF['reading_score'] >= 70) & 
                     (schoolDF["school_name"] == schoolData2["school_name"][x]) &
                                                              (schoolDF['math_score'] >= 70)]))
schoolData2["passed math and reading"] = students_passing_math_and_reading
schoolData2
```

```python
# Use the provided code to calculate the passing rates
schoolData2["% passed math"] = round(schoolData2['passed math'] / schoolData2['size'] * 100, 2)
schoolData2["% passed reading"] = round(schoolData2['passed reading'] / schoolData2['size']* 100, 2)
schoolData2["% passed math and reading"] = round(schoolData2['passed math and reading'] / schoolData2['size']* 100, 2)
schoolData2
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary2 = schoolData2[['school_name',"type", "size", "budget", "Per Capita Budget",
                                  "% passed math", "% passed reading" ,"% passed math and reading",
                                  "average math score","average reading score","passed math and reading"]]
per_school_summary = per_school_summary2.rename(columns= {"school_name":"School Name","budget":"Total School Budget",
                                                         "Per Capita Budget":"Per Student Budget",
                                                          "% passed math and reading":"% Overall Passing",
                                                         "size":"Students","% passed math":"% Math Passing",
                                                         "% passed reading":"% Reading Passing",
                                                         "average reading score":"Average Reading Score",
                                                         "average math score":"Average Math Score",
                                                         "passed math and reading":"Passed Math and Reading",
                                                         "type":"School Type"})
# Formatting
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

# Display the DataFrame
per_school_summary
```

## Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
per_school_summary
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = schoolData2.copy()
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per Capita Budget"],bins=spending_bins,
                                                            labels=labels, include_lowest=True)
school_spending_df
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["average math score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["average reading score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% passed math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% passed reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% passed math and reading"].mean()
overall_passing_spending
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame({"Spending Math Score":spending_math_scores,
                                 "Spending Reading Score":spending_reading_scores,
                                 "Spending Passing Math Score":spending_passing_math,
                                 "Spending Passing Reading Score":spending_passing_reading,
                                 "Spending Passing Overall Score":overall_passing_spending})

# Display results
spending_summary
```
